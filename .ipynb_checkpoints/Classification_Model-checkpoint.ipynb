```python
import numpy as np
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from collections import Counter
from sklearn import neighbors, svm, model_selection, metrics
from sklearn.ensemble import VotingClassifier, RandomForestClassifier
from sklearn.model_selection import TimeSeriesSplit, cross_validate, train_test_split
from sklearn.utils import resample
from tqdm import tqdm
import datetime as dt
```

```python
pd.set_option("display.max_columns", 15)
pd.set_option("display.width", 500)

# Variables
window_size = 3
moving_window = 15
```

```python
def classification(df):
    df['Next_change'] = df['Change'].shift(-1)
    df.fillna(0, inplace=True)
    df['Price_target'] = np.where(df['Next_change'] > 0.0, 1, -1)
    df.drop(columns='Next_change', inplace=True)
    
    print(df.head(10))


def data_processing(ticker):

    df = pd.read_csv('Data/EUR_USD_1418.csv', index_col=0)
        
    # Create labels for training
    classification(df)
    
    df.drop(columns=['Open','High','Low','Change'], inplace=True)
#     print(df.head())
    tickers = df.columns.values.tolist()
    df.fillna(0, inplace=True)

    # Compute values such that we get the percentage change of each values in each window from the start of the window.
#     for i in range(1, window_size+1):
#         df['{}_{}d'.format(ticker, i)] = ((df[ticker]) - df[ticker].shift(i)) / df[ticker].shift(i)
    
    # Percentage change between consecutive days.
    for i in range(1, window_size+1):
        df['{}_{}d'.format(ticker, i)] = ((df[ticker].shift(i-1)) - df[ticker].shift(i)) / df[ticker].shift(i)

    df.fillna(0, inplace=True)
    
    return tickers, df, window_size
```

```python
def feature_extraction(ticker):
    tickers, df, window_size = data_processing(ticker)
    
    tmp_df = df.copy()
    tmp_df.drop(columns=['Price', 'Price_target'], inplace=True)

    # Print number of 1, -1 and 0
    vals = df['{}_target'.format(ticker)].values.tolist()
    print('Data spread: ', Counter(vals))

    # Fill missing values
    df.fillna(0, inplace=True)
    df = df.replace([np.inf, -np.inf], np.nan)
    # Drop na, nan and nat rows
    df.dropna(inplace=True)
    
    df = df[5:]
    
    print(df.head(11))
    print(df.tail(11))
    
    tmp_df = df.copy()
    tmp_df.drop(columns=['Price', 'Price_target'], inplace=True)

    print(tmp_df.head())
    
    X = tmp_df.values
    y = df['{}_target'.format(ticker)].values
    print(X)
    print(y)

    return X, y, df


def machine_learning(ticker):
    
    start_time = dt.datetime.now()

    # X is feature sets
    # y is label sets
    # df is DataFrame
    X, y, df = feature_extraction(ticker)

    print("Original feature sets shape is {}".format(X.shape))
    print("Original label sets shape is {}".format(y.shape))

    # X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
    # print(X_train.shape)
    # print(X_test.shape)
    # print(y_train.shape)
    # print(y_test.shape)
    
    split_pct = 0.8
    train_size = int(X.shape[0] * split_pct)
    
    X_train, X_test = X[0:train_size], X[train_size:]
    y_train, y_test = y[0:train_size], y[train_size:]

#     print(X_train.shape)
#     print(y_train.shape)
#     print(X_test.shape)
#     print(y_test.shape)

    clf1 = neighbors.KNeighborsClassifier()
    clf2 = svm.SVC(kernel='linear', C=10)
    clf3 = RandomForestClassifier()
    clf4 = VotingClassifier([('lsvc', svm.LinearSVC()),
                            ('knn', neighbors.KNeighborsClassifier()),
                            ('rfor', RandomForestClassifier())])

#     print(y_train)

#     If there is only one classification, then we will automatically set the test data to that classification and
#     then validate it with its actual result.
    clf1.fit(X_train, y_train)
    accuracy = clf1.score(X_test, y_test)
    prediction = clf1.predict(X_test) 
    print('KNN Accuracy:', accuracy)
    
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, prediction).ravel()
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    
    f1_score = 2 * (precision * recall) / (precision + recall)
    print('f1_score:', f1_score)
    
    print("Confusion matrix: ")
    print(metrics.confusion_matrix(y_test, prediction))
    
    #
    clf2.fit(X_train, y_train)
    accuracy = clf2.score(X_test, y_test)
    prediction = clf2.predict(X_test)
    print('SVM Linear Accuracy:', accuracy)
    
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, prediction).ravel()
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    
    f1_score = 2 * (precision * recall) / (precision + recall)
    print('f1_score:', f1_score)
    
    print("Confusion matrix: ")
    print(metrics.confusion_matrix(y_test, prediction))
    
    #
    clf3.fit(X_train, y_train)
    accuracy = clf3.score(X_test, y_test)
    prediction = clf3.predict(X_test) 
    print('Random Forest Classification Accuracy:', accuracy)
    
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, prediction).ravel()
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    
    f1_score = 2 * (precision * recall) / (precision + recall)
    print('f1_score:', f1_score)
    
    print("Confusion matrix: ")
    print(metrics.confusion_matrix(y_test, prediction))
    
    #
    clf4.fit(X_train, y_train)
    accuracy = clf4.score(X_test, y_test)
    prediction = clf4.predict(X_test) 
    print('Pipeline Accuracy:', accuracy)
            
    tn, fp, fn, tp = metrics.confusion_matrix(y_test, prediction).ravel()
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
    
    f1_score = 2 * (precision * recall) / (precision + recall)
    print('f1_score:', f1_score)
    
    print("Confusion matrix: ")
    print(metrics.confusion_matrix(y_test, prediction))

#     print("Confusion matrix: ")
#     print(metrics.confusion_matrix(y_test, prediction))
    
    
warnings.filterwarnings(module='sklearn*', action='ignore', category=DeprecationWarning)
# process_data_for_labels('Close')
# feature_extraction('Close')
machine_learning('Price')

# To loop Through to find decent feature sets to use.
# for i in range(4):
#     window_size += 5
#     moving_window += 5
#     print('Processing windows size of {} and {}'.format(window_size, moving_window))
#     machine_learning('Close')
```
